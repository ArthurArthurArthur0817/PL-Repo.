```python
for x in soup.find_all('li', class_='List(n)'):
  print(x.text.strip(),"")
```

```python
# 建立空陣列

num = []
price = []
up_down = []
Quote_change = []
Highest = []
lowest = []
price_gap = []
Trading_volume = []
Transaction_amount = []


data = {
    "股名股號": num,
    "股價": price,
    "漲跌": up_down,
    "漲跌幅": Quote_change,
    "最高": Highest,
    "最低": lowest,
    "價差": price_gap,
    "成交量": Trading_volume,
    "成交金額": Transaction_amount
}
```

```python
i=0
for x in soup.find_all('div', class_='D(f) Ai(c)'):   #股號/股名
  if i!=0:
    num.append(x.text.strip())
  i+=1
print(num)
print(len(num))
```

```python

# 使用列表形式傳遞多個類別名稱
for x in soup.find_all('span', class_=['Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-up)', 'Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-down)', 'Jc(fe) Fw(600) C(#fff) Px(6px) Py(2px) Bdrs(4px) Bgc($c-trend-up)','Jc(fe) Fw(600) D(f) Ai(c)']):   #股價
    price.append(x.text.strip())

print(price)
print(len(price))


#Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-up) 正常紅色
#Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-down) 正常綠色
#Jc(fe) Fw(600) C(#fff) Px(6px) Py(2px) Bdrs(4px) Bgc($c-trend-up) 特殊紅色
#Jc(fe) Fw(600) D(f) Ai(c) 黑色
```

```python
i=0
for x in soup.find_all('span', class_=['Fw(600) Jc(fe) D(f) Ai(c) C($c-trend-up)','Fw(600) Jc(fe) D(f) Ai(c) C($c-trend-down)','Fw(600) Jc(fe) D(f) Ai(c)']):   #漲跌/漲跌幅
  if i%2==0:
    up_down.append(x.text.strip())
  else:
    Quote_change.append(x.text.strip())
  i+=1
print(up_down)
print(Quote_change)
print(len(up_down))
print(len(Quote_change))
#Fw(600) Jc(fe) D(f) Ai(c) C($c-trend-up) 正常紅色
#Fw(600) Jc(fe) D(f) Ai(c) C($c-trend-down)正常綠色
#Fw(600) Jc(fe) D(f) Ai(c) 黑色
```

```python
i=0
for x in soup.find_all('div', class_='Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend($m-table-cell-space) Mend(0):lc Miw(74px)'):  #較高/較低
  if i > 1:
    if i%2==0:
      Highest.append(x.text.strip())
    else:
      lowest.append(x.text.strip())
  i+=1
print(Highest)
print(lowest)
print(len(Highest))
print(len(lowest))
```

```python
i=0
for x in soup.find_all('div', class_='Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend($m-table-cell-space) Mend(0):lc Miw(52px)'):  #價差
  if i!=0:
    price_gap.append(x.text.strip())
  i+=1
print(price_gap)
print(len(price_gap))
```

```python
i=0
for x in soup.find_all('div', class_='Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend($m-table-cell-space) Mend(0):lc Miw(80px)'):  #成交量
  if i!=0:
     Trading_volume.append(x.text.strip())
  i+=1
print(Trading_volume)
print(len(Trading_volume))
```

```python
i=0
for x in soup.find_all('div', class_='Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend(0):lc Miw(88px) Mend(12px)'):  #成交量
  if i!=0:
     Transaction_amount.append(x.text.strip())
  i+=1
print(Transaction_amount)
print(len(Transaction_amount))
```

```python
# 指定要爬取的網址
url = link

# 發送 HTTP GET 請求，獲取網頁內容
response = requests.get(url)

# 檢查請求是否成功
if response.status_code == 200:
    # 使用 BeautifulSoup 解析 HTML 內容
    soup = BeautifulSoup(response.content, "html.parser")

    from bs4 import BeautifulSoup

      # 假設 soup 是你的 BeautifulSoup 物件
      # 找到所有的 <p> 標籤元素
    paragraphs = soup.find_all("p")

      # 找到所有的 <u> 標籤元素
    underline_texts = soup.find_all("h3")
    x=0
    y=0
      # 輸出找到的 <p> 標籤元素的文字內容
    for paragraph in paragraphs:
        print(paragraph.text)
        x+=1;

      # 輸出找到的 <u> 標籤元素的文字內容
    for underline_text in underline_texts:
        print(underline_text.text)
        y+=1
    print(x,y)
else:
    print("無法獲取網頁內容，HTTP 狀態碼：", response.status_code)
```

```python
# 建立一個空串列用來存放找到的標題內容
extracted_titles = []

h3_elements = soup.find_all("h3")

# 將找到的 <h3> 標籤元素的文字內容加入到 extracted_titles 串列中
for h3_element in h3_elements:
    extracted_titles.append(h3_element.text)

# 輸出 extracted_titles 串列中的內容
for title in extracted_titles:
    print(title)
```

```python
# 建立一個空串列用來存放找到的標題內容
extracted_content = []

p_elements = soup.find_all("p")

i=0
# 將找到的 <h3> 標籤元素的文字內容加入到 extracted_titles 串列中
for p_element in p_elements:
  if i>1:
    extracted_content.append(p_element.text)
  i+=1

# 輸出 extracted_titles 串列中的內容
for content in extracted_content:
    print(content)
```

```python
# 假設你已經有了兩個 list extracted_content 和 extracted_title

# 將兩個 list 合併成一個字典，其中 extracted_title 作為鍵，extracted_content 作為值
data = {"Title": extracted_titles ,"Content": extracted_content}

# 使用 DataFrame 构造函数將字典轉換為 DataFrame
df = pd.DataFrame(data)

# 印出 DataFrame
print(df)
```
