```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url ='https://tw.stock.yahoo.com/rank/volume?exchange=TAI'
```

```python
print(url)
```

```python
response = requests.get(url)
```

```python
response.text
```

```python
if response.status_code == 200:
  soup = BeautifulSoup(response.text, 'html.parser')
```

```python
for x in soup.find_all('li', class_='List(n)'):
  print(x.text.strip()," ")
```

```python
# 建立空陣列

num = []
price = []
up_down = []
Quote_change = []
Highest = []
lowest = []
price_gap = []
Trading_volume = []
Transaction_amount = []


data = {
    "股名股號": num,
    "股價": price,
    "漲跌": up_down,
    "漲跌幅": Quote_change,
    "最高": Highest,
    "最低": lowest,
    "價差": price_gap,
    "成交量": Trading_volume,
    "成交金額": Transaction_amount
}
```

```python
i=0
for x in soup.find_all('div', class_='D(f) Ai(c)'):   #股號/股名
  if i!=0:
    num.append(x.text.strip())
  i+=1
print(num)
print(len(num))
```

```python

# 使用列表形式傳遞多個類別名稱
for x in soup.find_all('span', class_=['Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-up)', 'Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-down)', 'Jc(fe) Fw(600) C(#fff) Px(6px) Py(2px) Bdrs(4px) Bgc($c-trend-up)','Jc(fe) Fw(600) D(f) Ai(c)']):   #股價
    price.append(x.text.strip())

print(price)
print(len(price))


#Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-up) 正常紅色
#Jc(fe) Fw(600) D(f) Ai(c) C($c-trend-down) 正常綠色
#Jc(fe) Fw(600) C(#fff) Px(6px) Py(2px) Bdrs(4px) Bgc($c-trend-up) 特殊紅色
#Jc(fe) Fw(600) D(f) Ai(c) 黑色
```

```python
i=0
for x in soup.find_all('span', class_=['Fw(600) Jc(fe) D(f) Ai(c) C($c-trend-up)','Fw(600) Jc(fe) D(f) Ai(c) C($c-trend-down)','Fw(600) Jc(fe) D(f) Ai(c)']):   #漲跌/漲跌幅
  if i%2==0:
    up_down.append(x.text.strip())
  else:
    Quote_change.append(x.text.strip())
  i+=1
print(up_down)
print(Quote_change)
print(len(up_down))
print(len(Quote_change))
#Fw(600) Jc(fe) D(f) Ai(c) C($c-trend-up) 正常紅色
#Fw(600) Jc(fe) D(f) Ai(c) C($c-trend-down)正常綠色
#Fw(600) Jc(fe) D(f) Ai(c) 黑色
```

```python
i=0
for x in soup.find_all('div', class_='Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend($m-table-cell-space) Mend(0):lc Miw(74px)'):  #較高/較低
  if i > 1:
    if i%2==0:
      Highest.append(x.text.strip())
    else:
      lowest.append(x.text.strip())
  i+=1
print(Highest)
print(lowest)
print(len(Highest))
print(len(lowest))
```

```python
i=0
for x in soup.find_all('div', class_='Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend($m-table-cell-space) Mend(0):lc Miw(52px)'):  #價差
  if i!=0:
    price_gap.append(x.text.strip())
  i+=1
print(price_gap)
print(len(price_gap))
```

```python
i=0
for x in soup.find_all('div', class_='Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend($m-table-cell-space) Mend(0):lc Miw(80px)'):  #成交量
  if i!=0:
     Trading_volume.append(x.text.strip())
  i+=1
print(Trading_volume)
print(len(Trading_volume))
```

```python
i=0
for x in soup.find_all('div', class_='Fxg(1) Fxs(1) Fxb(0%) Ta(end) Mend(0):lc Miw(88px) Mend(12px)'):  #成交量
  if i!=0:
     Transaction_amount.append(x.text.strip())
  i+=1
print(Transaction_amount)
print(len(Transaction_amount))
```

```python
print(len(num))
print(len(price))
print(len(up_down))
print(len(Quote_change))
print(len(Highest))
print(len(lowest))
print(len(price_gap))
print(len(Trading_volume))
print(len(Transaction_amount))
```

```python
df = pd.DataFrame(data)

df.index = [str(i) for i in range(1, 101)]  #自訂索引值
df.columns = ["Stock_Number", "price", "up_down","Quote_change","Highest","lowest","price_gap","Trading_volume","Transaction_amount(100 million)"]  #自訂欄位名稱

print(df)
# 將 DataFrame 輸出為 CSV 文件
df.to_csv('data.csv', index=False)
# 將 DataFrame 輸出為 JSON 文件
df.to_json('data.json', orient='records')

```
